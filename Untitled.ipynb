```python
words_count = {}
for line in y[0]:
    words = line.split(' ')
    for word in words:
        if word in words_count:
            words_count[word] += 1
        else:
            words_count[word] = 1
words_count
```

```python
sorted_words = sorted(words_count, key=words_count.get, reverse=True)
sorted_words
```
